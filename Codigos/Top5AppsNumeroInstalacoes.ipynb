```python
import pandas as pd
import matplotlib.pyplot as plt

# Leitura do arquivo CSV
database = "googleplaystore.csv"
dados = pd.read_csv(database, sep=',')

# Remoção de duplicatas
dados = dados.drop_duplicates()

# Verifica quais linhas da coluna 'Installs' começam com um número
filtro_installs_validos = dados['Installs'].str.contains(r'^\d', regex=True)

# Aplica o filtro para manter apenas as linhas válidas
dados = dados[filtro_installs_validos]

# Limpeza da coluna 'Installs': remove '+' e ',' e converte para inteiro
installs_limpo = dados['Installs'].str.replace('[+,]', '', regex=True).astype(int)
dados['Installs'] = installs_limpo

# Seleciona os 5 apps com mais instalações
top_5_apps = dados.nlargest(5, 'Installs')

# Define cores para o gráfico
cores = ["#b3ecba", "#73af7a", "#196f26", "#1B4E21", "#0a280d"]

# Criação do gráfico
plt.figure(figsize=(10, 6))
plt.barh(top_5_apps['App'], top_5_apps['Installs'], color=cores, height=0.6)

# Rótulos e título
plt.ylabel('App')
plt.xlabel('Número de Instalações')
plt.title('Top 5 Apps por Número de Instalações')

# Inverte eixo Y para maior valor no topo
plt.gca().invert_yaxis()

# Exibe o gráfico
plt.show()

```
