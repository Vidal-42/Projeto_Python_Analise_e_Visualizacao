```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregamento do arquivo CSV
df = pd.read_csv('googleplaystore.csv')

# Ordenao do DataFrame pelo rating em ordem decrescente
df_sorted = df.sort_values('Rating', ascending=False)

# Seleo dos 5 primeiros aplicativos com maior rating
top_5_apps = df_sorted.head(5)

colors = ['lightcoral', 'lightblue', 'lightgreen', 'plum', 'peachpuff']

# Criao do grfico 
plt.figure(figsize=(10, 6))
plt.plot(top_5_apps['App'], top_5_apps['Rating'], color='black', linestyle='-', alpha=0.7, linewidth=2)  

for i, app in enumerate(top_5_apps['App']):
    plt.scatter(app, top_5_apps['Rating'].iloc[i], color=colors[i], s=120, edgecolor='black', zorder=3)  

plt.title('Top 5 Apps com Maiores Ratings')
plt.xlabel('Aplicativo')
plt.ylabel('Rating')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
