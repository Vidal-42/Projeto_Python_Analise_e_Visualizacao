```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregamento do arquivo CSV 'googleplaystore.csv' no DataFrame 'dataset' e remoção de linhas duplicadas
dataset = pd.read_csv('googleplaystore.csv', sep=',') 
dataset = dataset.drop_duplicates()

# Padronização da coluna 'Installs', com remoção de símbolos de '+' e ',' para facilitar a conversão para numérico
dataset['Installs'] = dataset['Installs'].str.replace('+', '', regex=False).str.replace(',', '', regex=False)
dataset['Installs'] = pd.to_numeric(dataset['Installs'], errors='coerce')  # Conversão da coluna 'Installs' para o tipo numérico, com tratamento de erros

# Remoção de linhas com valor nulo na coluna 'Installs'
dataset = dataset.dropna(subset=['Installs'])

# Seleção dos 10 aplicativos com o menor número de instalações, em ordem crescente
top_10_installs = dataset[['App', 'Installs']].sort_values(by='Installs', ascending=True).head(10)

# Definição de uma paleta de cores as bolhas do gráfico de dispersão
vibrant_colors = ['#FF5733', '#FF8D1A', '#FFC300', '#DAF7A6', '#33FF57', '#33FFBD', '#3380FF', '#9D33FF', '#FF33A6', '#FF3333']

# Criação da figura do gráfico de dispersão com o tamanho especificado
plt.figure(figsize=(10, 6))

# Plotagem do gráfico de dispersão com cores e tamanho dos círculos aumentados (s=100)
plt.scatter(top_10_installs['App'], top_10_installs['Installs'], color=vibrant_colors, s=100)

# Adição de título e rótulos dos eixos
plt.title('Top 10 Aplicativos com Menor Número de Instalações')
plt.xlabel('Aplicativos')
plt.ylabel('Número de Instalações')

# Rotação dos rótulos do eixo x 
plt.xticks(rotation=45)

# Exibição da grade no gráfico 
plt.grid()

# Ajuste do layout para evitar sobreposição dos elementos
plt.tight_layout()

# Exibição do gráfico
plt.show()


```
