```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregamento do arquivo CSV 'googleplaystore.csv' no DataFrame 'dataset' e remoo de linhas duplicadas
dataset = pd.read_csv('googleplaystore.csv', sep=',') 
dataset = dataset.drop_duplicates()

# Padronizao da coluna 'Installs', com remoo de smbolos de '+' e ',' para facilitar a converso para numrico
dataset['Installs'] = dataset['Installs'].str.replace('+', '', regex=False).str.replace(',', '', regex=False)
dataset['Installs'] = pd.to_numeric(dataset['Installs'], errors='coerce')  # Converso da coluna 'Installs' para o tipo numrico, com tratamento de erros

# Remoo de linhas com valor nulo na coluna 'Installs'
dataset = dataset.dropna(subset=['Installs'])

# Seleo dos 10 aplicativos com o menor nmero de instalaes, em ordem crescente
top_10_installs = dataset[['App', 'Installs']].sort_values(by='Installs', ascending=True).head(10)

# Definio de uma paleta de cores as bolhas do grfico de disperso
vibrant_colors = ['#FF5733', '#FF8D1A', '#FFC300', '#DAF7A6', '#33FF57', '#33FFBD', '#3380FF', '#9D33FF', '#FF33A6', '#FF3333']

# Criao da figura do grfico de disperso com o tamanho especificado
plt.figure(figsize=(10, 6))

# Plotagem do grfico de disperso com cores e tamanho dos crculos aumentados (s=100)
plt.scatter(top_10_installs['App'], top_10_installs['Installs'], color=vibrant_colors, s=100)

# Adio de ttulo e rtulos dos eixos
plt.title('Top 10 Aplicativos com Menor Nmero de Instalaes')
plt.xlabel('Aplicativos')
plt.ylabel('Nmero de Instalaes')

# Rotao dos rtulos do eixo x 
plt.xticks(rotation=45)

# Exibio da grade no grfico 
plt.grid()

# Ajuste do layout para evitar sobreposio dos elementos
plt.tight_layout()

# Exibio do grfico
plt.show()


```
