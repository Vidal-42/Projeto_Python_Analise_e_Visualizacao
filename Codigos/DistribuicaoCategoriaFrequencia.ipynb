```python
import pandas as pd
import matplotlib.pyplot as plt

# Leitura do arquivo csv
database = 'googleplaystore.csv'
dataframe = pd.read_csv(database)

# Remoção de duplicatas
arquivo = dataframe.drop_duplicates()

categoria_frequencia = dataframe['Category'].value_counts()

plt.figure(figsize=(18, 11))
plt.pie(categoria_frequencia, labels=categoria_frequencia.index, autopct='%1.1f%%', startangle=900, textprops={'fontsize': 6})
plt.title('Distribuição de Aplicativos por Categoria e Frequências de aparecimento')
plt.axis('equal') 
plt.show()

```
